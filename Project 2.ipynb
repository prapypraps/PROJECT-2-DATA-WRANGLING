```python
# working on outliers
# We will do outlier treatment for temp , atemp , hum , windspeed

def wisker (col):
    q1,q3=np.percentile(col,[25,75])
    iqr=q3-q1
    lw=q1-1.5*iqr
    uw=q3+1.5*iqr
    return lw,uw
```

```python
wisker(df['temp'])
```

```python
for i in ['temp','atemp','hum','windspeed']:
    lw,uw=wisker(df[i])
    df[i]=np.where(df[i]<lw,lw,df[i])
    df[i]=np.where(df[i]>uw,uw,df[i])
                
```

```python
for i in ['temp','atemp','hum','windspeed']: # checking wether it is resolved or now
    sns.boxplot(df[i])
    plt.show()
```

```python
# correlation with heat map to interpret relationship

s=df.select_dtypes(include='number').corr()
```
